## Запрос № 1 к ChatGPT:

Напиши подготовку тренировочных данных для нейронной сети по классификации изображений, для этого для напиши код скачивания  датасета при помощи gdown по этой ссылке 'https://drive.google.com/uc?id=17g43gjghHXxYv_slVNhfU8p8Yqw8jew1', и код разархивации датасета из zip формата в директорию '/content/reality'. Датасет не поделен на train и test. Он состоит из 2 папок, название каждой папки соответствует названию категории, каждый файл необходимо открыть при помощи Image.open, не используй генератор. Учти, что изображения черно-белые. Добавь в  код масштабирование данных. Используй LabelEncoder для кодирования строковых меток в целые числа.  Учти что изображения черно-белые. Новый размер изображений должен быть 64 на 64. Раздели на train и test (80% train, 20% test).  Определи список названий классов и количество классов. Напиши количество изображений для каждого класса.

>Ответ ChatGPT:

````python
import os
import gdown
from zipfile import ZipFile
from PIL import Image
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import numpy as np

# Задаем путь к zip-архиву и директории, куда будем извлекать данные
zip_path = '/content/dataset.zip'
extract_path = '/content/reality'

# Скачиваем датасет при помощи gdown
gdown.download('https://drive.google.com/uc?id=17g43gjghHXxYv_slVNhfU8p8Yqw8jew1', zip_path, quiet=False)

# Разархивируем датасет
with ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Определяем пути к изображениям и их меткам
image_paths = []
labels = []

for category in os.listdir(extract_path):
    category_path = os.path.join(extract_path, category)
    if os.path.isdir(category_path):
        for image_file in os.listdir(category_path):
            image_path = os.path.join(category_path, image_file)
            image_paths.append(image_path)
            labels.append(category)

# Преобразуем строки меток в целые числа с использованием LabelEncoder
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Загружаем изображения, масштабируем и изменяем размер на 64x64
images = []
for image_path in image_paths:
    image = Image.open(image_path).convert('L')  # 'L' - режим черно-белого изображения
    image = image.resize((64, 64))
    image_array = np.array(image) / 255.0  # Масштабирование данных
    images.append(image_array)

# Разделяем данные на тренировочный и тестовый наборы (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(images, encoded_labels, test_size=0.2, random_state=42)

# Определяем список названий классов и количество классов
class_names = label_encoder.classes_
num_classes = len(class_names)

# Выводим количество изображений для каждого класса
for class_name in class_names:
    class_count = labels.count(class_name)
    print(f'Class: {class_name}, Count: {class_count}')

# Теперь у вас есть данные для тренировки и тестирования нейронной сети.
# X_train, y_train - тренировочный набор данных
# X_test, y_test - тестовый набор данных
# class_names - список названий классов
# num_classes - количество классов
````

## Запрос № 2 к ChatGPT:

Отлично получилось! Теперь напиши еще один код для гугл колаба, который бы выводил  по три примера изображений для каждого класса. Подпиши названия классов для изображений

>Ответ ChatGPT:

````python
import matplotlib.pyplot as plt

# Функция для вывода изображений для каждого класса
def display_images_per_class(images, labels, class_names):
    fig, axs = plt.subplots(len(class_names), 3, figsize=(10, 15))
    fig.subplots_adjust(hspace=0.5)

    for i, class_name in enumerate(class_names):
        class_indices = [j for j, label in enumerate(labels) if label == i][:3]

        for j, index in enumerate(class_indices):
            axs[i, j].imshow(images[index], cmap='gray')
            axs[i, j].axis('off')

            if j == 0:
                axs[i, j].set_title(f'Class: {class_name}')

    plt.show()

# Выводим по три примера изображений для каждого класса
display_images_per_class(X_train, y_train, class_names)

````

````python
#@title Формирование X_train, y_train, X_test, y_test

# Создаем пустые списки для хранения путей к файлам изображений и меток классов
image_paths = []
labels = []

# Получаем список папок, содержащихся в директории 'construction'
dir_path = 'construction'
categories = os.listdir(dir_path)

# Проходим по каждой папке и добавляем в списки путь к файлу изображения и метку класса
for category in categories:
    category_path = os.path.join(dir_path, category)
    if os.path.isdir(category_path):
        files = os.listdir(category_path)
        #print(f"Number of images in '{category}': {len(files)}")
        for file_name in files:
            file_path = os.path.join(category_path, file_name)
            image_paths.append(file_path)
            labels.append(category)

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

# Создаем экземпляр LabelEncoder и преобразуем лейблы в числовой формат
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(labels)

# Создаем экземпляр OneHotEncoder и преобразуем числовые значения в One-Hot Encoding
onehot_encoder = OneHotEncoder(sparse=False)
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)

import numpy as np
from keras.utils import to_categorical

# Создаем пустой список для изображений
images = []

# Проходим по каждому пути к файлу изображения и добавляем его в список
for path in image_paths:
    with Image.open(path) as img:
        img = img.resize((64, 64)) # изменяем размер изображения до 64x64 пикселей
        img = np.array(img) # приводим изображение к формату numpy массива
        images.append(img) # добавляем изображение в список

# Преобразуем список в формат numpy массива
images = np.array(images)

# Преобразуем список меток классов в формат numpy массива

labels_onehot = np.array(onehot_encoded)

# Нормируем массив изображений
images_normalized = images / 255.0

from sklearn.model_selection import train_test_split

# Разделяем данные на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(images, labels_onehot, test_size=0.2, random_state=42)

# Выводим размеры тренировочной и тестовой выборок
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
````
